```python
import pandas as pd
import requests
import time

API_KEY = 'cb930dd34a5901475d162cb4788ea98d'

df = pd.read_csv('data/viewing_releases.csv')

results = []

def get_tmdb_data(imdb_id):
    url = f"https://api.themoviedb.org/3/find/{imdb_id}?api_key={API_KEY}&external_source=imdb_id"

    try:
        response = requests.get(url)
        data = response.json()

        if data['movie_results']:
            return 'movie'

        elif data['tv_results']:
            return 'tv'

        else:
            return None

    except:
        return None

for i, imdb_id in enumerate(df['imdb_id'], start=1):
    media_type = get_tmdb_data(imdb_id)

    results.append({
        'imdb_id': imdb_id,
        'media_type': media_type
    })

    print(f"{i}/{len(df)} - {imdb_id} -> {media_type}")

    time.sleep(0.5)

results_df = pd.DataFrame(results)
results_df.to_csv('data/tmdb_data.csv', index = False)
print('Klaar, CSV is opgeslagen')
    
```
