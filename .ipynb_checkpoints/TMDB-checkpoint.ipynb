```python
import pandas as pd
import requests
import time

API_KEY = 'cb930dd34a5901475d162cb4788ea98d'

df = pd.read_csv('viewing_releases.csv')

results = []

def get_tmdb_data(imdb_id):
    url = f"https://api.themoviedb.org/3/find/{imdb_id}?api_key={API_KEY}&external_source=imdb_id"

    try:
        response = requests.get(url)
        data = response.json()

        if data['movie_results']:
            result = data['movie_results'][0]
        elif data ['tv_results']:
            result = data['tv_results'][0]
        else:
            return none, none

        return result.get('popularity'), result.get('media_type')

    except:
        return none, none

for imdb_id in df['imdb_id']:
    popularity, media_type = get_tmdb_data(imdb_id)

    results.append({
        'imdb_id': imdb_id,
        'tmdb_popularity': popularity,
        'media_type': media_type
    })

    time.sleep(0.5)

results_df = pd.DataFrame(results)
results_df.to_csv('tmdb_data.csv', index = False)
print('Klaar, CSV is opgeslagen')
    
```
